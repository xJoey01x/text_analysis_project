## Importing Libraries

```python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
import nltk
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import string
import csv
```

## Reading the dataset

```python
df = pd.read_excel('D:\MY FOLDER\Input.xlsx')
df
```

## Scraping Title and Content from URLs

```python
link = [url for url in df['URL']]
# link
```

```python
title = []
content = []
title_count=0
content_count=0
counter=0

for url in link:
    counter+=1 
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")
    
    
    #titles only
    try:
        title1 = soup.find('h1',attrs={'class': 'entry-title'})
        title1 = title1.text.replace('\n'," ")
        title.append(title1)
    #print(title1)
    except:
        title_count+=1
    
    
    #content only
    try:
        content1 = soup.find(attrs={'class':'td-post-content'})
        content1 = content1.text.replace('\n'," ")
        content.append(content1)
    #print(content1)
    except:
        content_count+=1
        print(counter)

print("Unable to fetch titles of "+str(title_count)+" articles.")
print("Unable to fetch content of "+str(content_count)+" articles.")
```

```python
#Dropping the rows for which content could not be fetched
index_to_drop = [36,49]

df = df.drop(index_to_drop)
df
```

## Converting articles to sentences

```python
nltk.download('punkt')

sentences = []

for article in content:
  sentences.append(len(sent_tokenize(article)))
```

```python
print(sentences)
```

## Removing Punctuations

```python
for i in range(len(content)):
    content[i] = content[i].translate(str.maketrans('', '', string.punctuation)) 
```

## Converting to Tokens

```python
text_tokens = []

for i in range(len(content)):
    text_tokens1 = word_tokenize(content[i])
    text_tokens.append(text_tokens1)
```

```python
#Seeing the length of articles
for i in range(len(text_tokens)):
    print(len(text_tokens[i]))
```

## Removing Stopwords

```python
import nltk
from nltk.corpus import stopwords

with open('D:\MY FOLDER\Stop Words\stopwordsfile.txt' ,'r') as stop_words:
    stopWords = stop_words.read().lower()

stopWordList = stopWords.split('\n')
```

```python
no_stop_tokens = []

for i in range(len(text_tokens)):
    filtered_text = [word for word in text_tokens[i] if word.lower() not in stopWordList]
    no_stop_tokens.append(filtered_text)
```

```python
# Length of text_tokens after removing stopwords

for i in range(0,5):
    print("Length of first article before removing Stop Words: ", len(text_tokens[i]))

    print("Length of first article after removing Stop Words: ", len(no_stop_tokens[i]))
```

## Positive Score

#### Check for Positive words

```python
with open("D:/MY FOLDER/positive-words.txt") as pos:
    poswords = pos.read().split("\n")
```

```python
POS_count = []

for tokens in no_stop_tokens:
    pos_count = [w for w in tokens if w in poswords]
    POS_count.append(pos_count)
```

#### Calculating Positive Score

```python
Positive_score = [len(tokens) for tokens in POS_count]
print(positive_score)
```

## Negative Score

#### Check for Negative words

```python
with open("D:/MY FOLDER/negative-words.txt") as neg:
    negwords = neg.read().split("\n")
```

```python
NEG_count = []

for tokens in no_stop_tokens:
    neg_count = [w for w in tokens if w in negwords]
    NEG_count.append(neg_count)
```

#### Calculating Negative Score

```python
Negative_score = [len(tokens) for tokens in NEG_count]
print(negative_score)
```

```python
filtered_content = [" ".join(tokens) for tokens in no_stop_tokens]
```

## Subjectivity Score

Subjectivity Score = (Positive Score + Negative Score) / ((Total Words after cleaning) + 0.000001)

```python
Subjectivity_score = []
for i in range(len(no_stop_tokens)):
    sub_score = round((Positive_score[i] + Negative_score[i])/ (len(no_stop_tokens[i]) + 0.000001),4)
    Subjectivity_score.append(sub_score)
    
print(Subjectivity_score)
```

## Polarity Score

Polarity Score = (Positive Score – Negative Score) / ((Positive Score + Negative Score) + 0.000001)

```python
Polarity_score = []

for i in range(len(no_stop_tokens)):
    pol_score = round((Positive_score[i] - Negative_score[i])/ ((Positive_score[i] + Negative_score[i]) + 0.000001),4)
    Polarity_score.append(pol_score)
    
print(Polarity_score)
```

## Average sentence length

This is the average number of words per sentence.

```python
Avg_sentence_len = []

for i in range(len(no_stop_tokens)):
    avg_sent_len = round(len(text_tokens[i])/(sentences[i]),4)
    Avg_sentence_len.append(avg_sent_len)
    
print(Avg_sentence_len)
```

## Complex Word Count

Complex words are words in the text that contain more than two syllables.

```python
sylabble_counts1 = []
complex_words1 = []

for article in filtered_content:
    
    sylabble_count1=0
    d=article.split()
    ans1=0
    for word in d:
        count1=0
        for i in range(len(word)):
          if(word[i]=='a' or word[i]=='e' or word[i] =='i' or word[i] == 'o' or word[i] == 'u'):
               count1+=1
    #            print(words[i])
          if(i==len(word)-2 and (word[i]=='e' and word[i+1]=='d')):
            count1-=1;
          if(i==len(word)-2 and (word[i]=='e' and word[i]=='s')):
            count1-=1;
    sylabble_count1+=count1    
    if(count1>2):
        ans1+=1
    sylabble_counts1.append(sylabble_count1)
    complex_words1.append(ans1)       
```

## Word Count

Total number of cleaned words present in the text.

```python
Word_Count = [len(tokens) for tokens in no_stop_tokens]
    
print(Word_Count)
```

## Percentage of Complex Words

Percentage of Complex words = the number of complex words / the number of words

```python
Pct_complex_words = []

for i in range(len(no_stop_tokens)):
    pcw = round((complex_words1[i]/Word_Count[i]) * 100,2)
    Pct_complex_words.append(pcw)
    
print(Pct_complex_words)
```

## Fog Index

Fog Index = 0.4 * (Average Sentence Length + Percentage of Complex words)

```python
Fog_Index = []

for i in range(len(no_stop_tokens)):
    fog_index = round(0.4 * (Avg_sentence_len[i] + Pct_complex_words[i]),4)
    Fog_Index.append(fog_index)
    
print(Fog_Index)
```

## Average Number of Words per Sentence

The formula for calculating is:
Average Number of Words Per Sentence = the total number of words / the total number of sentences

```python
Avg_words_per_sentence = Avg_sentence_len
print(Avg_words_per_sentence)
```

## Syllable per Word

Number of Syllables in each word of the text can be calculated by counting the vowels present in each word. We also handle some exceptions like words ending with "es","ed" by not counting them as a syllable.

```python
Syllable_per_word = []

for i in range(len(no_stop_tokens)):
    word = filtered_content[i].replace(" ","")
    syllable_count = 0
    for w in word:
        if(w=='a' or w=='e' or w=='i' or w=='o' or w=='y' or w=='u' or w=='A' or w=='E' or w=='I' or w=='O' or w=='U' or w=='Y'):
            syllable_count += 1
    
    syllable_p_word = round(syllable_count / len(filtered_content[i].split()),4)
    Syllable_per_word.append(syllable_p_word)

print(Syllable_per_word)
```

## Personal Pronouns

To calculate Personal Pronouns, we find the counts of the words - “I,” “we,” “my,” “ours,” and “us”.

```python
Personal_Nouns = []
p_noun =['i', 'we','my', 'ours','us'] 
for i in range(len(text_tokens)):
    ans=0
    for word in text_tokens[i]:
        if word.lower() in p_noun:
          ans+=1
    Personal_Nouns.append(ans)

print(Personal_Nouns)
```

## Average Word Length

Average Word Length is calculated by the formula:
Sum of the total number of characters in each word/Total number of words

```python
total_characters = []
for article in content:
    characters = 0
    for word in article.split():
        characters+=len(word)
    total_characters.append(characters)
```

```python
Avg_word_length = []

for i in range(len(text_tokens)):
    avg_word_len = round(np.array(total_characters[i])/len(text_tokens[i]),4)
    Avg_word_length.append(avg_word_len)
    
print(Avg_word_length)
```
