### Importing Libraries

```python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
import nltk
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import string
import csv
```

### Importing the data

```python
df = pd.read_excel('D:\Folder\Input.xlsx')
df
```

### Getting the title and content from the urls

```python
link = [url for url in df['URL']]
# link
```

```python
title = []
content = []
title_count=0
content_count=0

for url in link:
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")
    
    
    #titles only
    try:
        title1 = soup.find('h1',attrs={'class': 'entry-title'})
        title1 = title1.text.replace('\n'," ")
        title.append(title1)
    #print(title1)
    except:
        print("Unable to fetch title!")
        title_count+=1
    
    
    #content only
    try:
        content1 = soup.find(attrs={'class':'td-post-content'})
        content1 = content1.text.replace('\n'," ")
        content.append(content1)
    #print(content1)
    except:
        print("Unable to fetch content!")
        content_count+=1

print("Unable to fetch titles of "+str(title_count)+" articles.")
print("Unable to fetch content of "+str(content_count)+" articles.")
```

### Getting the sentences from articles

```python
sentences = []

for article in content:
  sentences.append(len(sent_tokenize(article)))
```

```python
print(sentences)
```

### Removing punctuation from the content

```python
for i in range(len(content)):
    content[i] = content[i].translate(str.maketrans('', '', string.punctuation)) 
#     print(content[i])
```

### Converting into tokens

```python
text_tokens = []

for i in range(len(content)):
    text_tokens1 = word_tokenize(content[i])
    text_tokens.append(text_tokens1)
#     print(text_tokens1)
```

```python
for i in range(len(text_tokens)):
    print(len(text_tokens[i]))
```

```python
import nltk
from nltk.corpus import stopwords

with open('D:\Folder\Stop Words\stopwordsfile.txt' ,'r') as stop_words:
    stopWords = stop_words.read().lower()

stopWordList = stopWords.split('\n')


no_stop_tokens = []

for i in range(len(text_tokens)):
    for word in text_tokens[i]:
    filtered_text = [word for word in text_tokens[i] if word.lower() not in stopWordList]
    no_stop_tokens.append(filtered_text)

# print(no_stop_tokens)
```

### Check for Positive words

```python
with open("D:/Folder/positive-words.txt") as pos:
    poswords = pos.read().split("\n")
```

```python
# poswords
```

```python
POS_count = []

for i in range(len(no_stop_tokens)):
    pos_count = " ".join([w for w in no_stop_tokens[i] if w in poswords])
    pos_count = pos_count.split(" ")
    POS_count.append(pos_count)
#     print(pos_count)
```

### Calculating Positive Score

```python
Positive_score = []

for i in range(len(POS_count)):
    pos_score = len(POS_count[i])
    Positive_score.append(pos_score)
    
   
print(Positive_score)
```

### Check for Negative words

```python
with open("D:/Folder/negative-words.txt") as neg:
    negwords = neg.read().split("\n")
```

```python
# negwords
```

```python
NEG_count = []

for i in range(len(no_stop_tokens)):
    neg_count = " ".join([w for w in no_stop_tokens[i] if w in negwords])
    neg_count = neg_count.split(" ")
    NEG_count.append(neg_count)
#     print(neg_count)
```

### Calculating Negative Score

```python
Negative_score = []

for i in range(len(NEG_count)):
    neg_score = len(NEG_count[i])
    Negative_score.append(neg_score)
    
   
print(Negative_score)
```

```python
filter_content = []

for i in range(len(no_stop_tokens)):
    filtered_article = " ".join(no_stop_tokens[i])
    filter_content.append(filtered_article)
    
# print(filter_content)
```

```python
len(filter_content)
```

### Subjectivity Score

```python
# Subjectivity Score = (Positive Score + Negative Score)/ ((Total Words after cleaning) + 0.000001)

Subjectivity_score = []
for i in range(len(no_stop_tokens)):
    sub_score = (Positive_score[i] + Negative_score[i])/ (len(no_stop_tokens[i]) + 0.000001)
    Subjectivity_score.append(sub_score)
    
print(Subjectivity_score)
```

### Polarity Score

```python
# Polarity Score = (Positive Score – Negative Score)/ ((Positive Score + Negative Score) + 0.000001)

Polarity_score = []

for i in range(len(no_stop_tokens)):
    pol_score = (Positive_score[i] - Negative_score[i])/ ((Positive_score[i] + Negative_score[i]) + 0.000001)
    Polarity_score.append(pol_score)
    
print(Polarity_score)
```

### Average sentence length

```python
Avg_sentence_len = []

for i in range(len(no_stop_tokens)):
    avg_sent_len = len(text_tokens[i])/(sentences[i])
    Avg_sentence_len.append(avg_sent_len)
    
print(Avg_sentence_len)
```

### Complex Word Count

```python
#Complex words are words in the text that contain more than two syllables.

sylabble_counts1 = []
complex_words1 = []

for article in filter_content:
  sylabble_count1=0
  d=article.split()
  ans1=0
  for word in d:
    count1=0
    for i in range(len(word)):
      if(word[i]=='a' or word[i]=='e' or word[i] =='i' or word[i] == 'o' or word[i] == 'u'):
           count1+=1
#            print(words[i])
      if(i==len(word)-2 and (word[i]=='e' and word[i+1]=='d')):
        count1-=1;
      if(i==len(word)-2 and (word[i]=='e' and word[i]=='s')):
        count1-=1;
    sylabble_count1+=count1    
    if(count1>2):
        ans1+=1
  sylabble_counts1.append(sylabble_count1)
  complex_words1.append(ans1)           
```

```python
print(sylabble_counts1)
```

### Word Count

```python
#Total number of cleaned words present in the text. 

Word_Count = []


for i in range(len(no_stop_tokens)):
    word_count = len(no_stop_tokens[i])
    Word_Count.append(word_count)
    
print(Word_Count)
```

### Percentage of complex words

```python
Pct_complex_words = []

for i in range(len(no_stop_tokens)):
    pcw = round((complex_words1[i]/Word_Count[i]) * 100,2)
    Pct_complex_words.append(pcw)
    
print(Pct_complex_words)
```

### Fog Index

```python
#Fog Index = 0.4 * (Average Sentence Length + Percentage of Complex words)

Fog_Index = []

for i in range(len(no_stop_tokens)):
    fog_index = 0.4 * (Avg_sentence_len[i] + Pct_complex_words[i])
    Fog_Index.append(fog_index)
    
print(Fog_Index)
```

### Average number of words per sentence

```python
Avg_words_per_sentence = Avg_sentence_len
print(Avg_words_per_sentence)
```

### Syllable per Word

```python
Syllable_per_word = []

for i in range(len(no_stop_tokens)):
    word = filter_content[i].replace(" ","")
    syllable_count = 0
    for w in word:
        if(w=='a' or w=='e' or w=='i' or w=='o' or w=='y' or w=='u' or w=='A' or w=='E' or w=='I' or w=='O' or w=='U' or w=='Y'):
            syllable_count += 1
    
    syllable_p_word = syllable_count / len(filter_content[i].split())
    Syllable_per_word.append(syllable_p_word)

print(Syllable_per_word)
```

### Personal Pronouns

```python
Personal_Nouns = []
p_noun =['i', 'we','my', 'ours','us'] 
for i in range(len(text_tokens)):
    ans=0
    for word in text_tokens[i]:
        if word.lower() in p_noun:
          ans+=1
    Personal_Nouns.append(ans)

print(Personal_Nouns)
```

### Average Word Length

```python
total_characters = []
for article in content:
  characters = 0
  for word in article.split():
    characters+=len(word)
  total_characters.append(characters)  
```

```python
Avg_word_length = []

for i in range(len(text_tokens)):
    avg_word_len = np.array(total_characters[i])/len(text_tokens[i])
    Avg_word_length.append(avg_word_len)
    
print(Avg_word_length)
```
